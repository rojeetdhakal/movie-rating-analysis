# 1. Top 10 rated movies according to Rotten Tomatoes?

```python
MovieRatings.rename (columns={"Rotten Tomatoes":"RottenTomatoes"},inplace=True)
```

```python
MovieRatings.head()
```

```python
MovieRatings["RottenTomatoes"]=MovieRatings.RottenTomatoes.apply(lambda x:x.replace("/100"," ")).astype("int64")
```

```python
MovieRatings.head()
```

```python
MovieRatings.RottenTomatoes.dtype
```

```python
MovieRatings.sort_values(by= "RottenTomatoes",ascending=False).Title.head(10)
```

# How many different  age catagories are there? And plot a graph between different categories of age vs no. of movies?

```python
agegroups=MovieRatings.Age.unique()
```

```python
agegroups
```

```python
agegroups=MovieRatings.groupby("Age").Title.count()
```

```python
agegroups
```

```python
agegroups.sum()
```

```python
plt.plot(agegroups)
plt.xlabel("Age Categories")
plt.ylabel("Number of Movies")
plt.show()
```
